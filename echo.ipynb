### opencv

```python
import cv2

cv2.__version__
print("OpenCV version:", cv2.__version__)
```

```python
# open cv default variables
FRAME_WIDTH = 640
FRAME_HEIGHT = 480
```

```python
# reference from rgb2orbcam.cpp
def try_open_camera(camera_index: int, cap: cv2.VideoCapture, first_frame: cv2.Mat) -> bool:
    cap.release()  # Release any previously opened camera
    first_frame.release()  # Release any previously captured frame

    if not cap.open(camera_index):
        return False
    
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, FRAME_WIDTH)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, FRAME_HEIGHT)
    if not cap.read(first_frame) or first_frame.empty():
        return False
    
    return True
```
